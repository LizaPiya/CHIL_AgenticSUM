```python
import pandas as pd

# =========================
# Load CSV
# =========================
df = pd.read_csv("Final.csv")

# =========================
# Ground truth (LESS hallucination)
# =========================
GROUND_TRUTH = {
    "Psychiatric": "B",
    "Cardiopulmonary": "A",
    "Oncology": "B",
}

# =========================
# Column mappings
# =========================
CONFIG = {
    "Psychiatric": {
        "choice_col": "Doc 1(Psychiatric) : Which summary has more hallucination?",
        "vanilla_col": "psych_level_opt1",
        "agentic_col": "psych_level_opt2",
    },
    "Cardiopulmonary": {
        "choice_col": "Doc 2(Cardiopulmonary) : Which summary has more hallucination?",
        "vanilla_col": "cardio_level_opt2",
        "agentic_col": "cardio_level_opt1",
    },
    # ✅ FIXED: Oncology mapping corrected so B (LESS hallucinated) = Agentic
    "Oncology": {
        "choice_col": "Doc 3(Oncology): Which summary has more hallucination?",
        "vanilla_col": "onco_level_opt1",
        "agentic_col": "onco_level_opt2",
    },
}

# =========================
# Compute results
# =========================
results = []

for domain, cfg in CONFIG.items():

    # Annotator answer: MORE hallucinated
    more_hallucinated = df[cfg["choice_col"]]

    # Convert to LESS hallucinated (single explicit inversion)
    inferred_less = more_hallucinated.map({"A": "B", "B": "A"})

    # Accuracy against ground truth (LESS hallucinated)
    correct = (inferred_less == GROUND_TRUTH[domain]).mean() * 100

    # Mean severity (pure aggregation)
    vanilla_mean = df[cfg["vanilla_col"]].mean()
    agentic_mean = df[cfg["agentic_col"]].mean()

    results.append({
        "Clinical Domain": domain,
        "Correct Guess Accuracy (%)": round(correct, 1),
        "Vanilla Mean Severity": round(vanilla_mean, 2),
        "AgenticSum Mean Severity": round(agentic_mean, 2),
    })

# =========================
# Final table
# =========================
results_df = pd.DataFrame(results)
print(results_df)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import wilcoxon, binomtest

# Load data
df = pd.read_csv("Final.csv")

# -----------------------------
# Helper: rank-biserial effect size (CORRECT for Wilcoxon)
# -----------------------------
def rank_biserial(delta):
    pos = np.sum(delta > 0)
    neg = np.sum(delta < 0)
    if pos + neg == 0:
        return 0.0
    return (pos - neg) / (pos + neg)

# =============================
# 1. SEVERITY TESTS (Wilcoxon)
# =============================

# Psychiatric: Vanilla > Agentic
delta_psych = df["psych_level_opt1"] - df["psych_level_opt2"]
w_psych, p_psych = wilcoxon(delta_psych, alternative="greater")
r_psych = rank_biserial(delta_psych)

# Cardiopulmonary: Vanilla > Agentic
delta_cardio = df["cardio_level_opt2"] - df["cardio_level_opt1"]
w_cardio, p_cardio = wilcoxon(delta_cardio, alternative="greater")
r_cardio = rank_biserial(delta_cardio)


# Oncology: A more hallucinated (Vanilla worse)
delta_onco = df["onco_level_opt1"] - df["onco_level_opt2"]
w_onco, p_onco = wilcoxon(delta_onco, alternative="greater")
r_onco = rank_biserial(delta_onco)


print("=== Paired Severity Tests (Wilcoxon) ===")
print(f"Psychiatric     | W={w_psych:.1f} | p={p_psych:.6f} | r_rb={r_psych:.3f}")
print(f"Cardiopulmonary | W={w_cardio:.1f} | p={p_cardio:.6f} | r_rb={r_cardio:.3f}")
print(f"Oncology        | W={w_onco:.1f} | p={p_onco:.6f} | r_rb={r_onco:.3f}")

# =================================
# 2. DOMINANCE / PREFERENCE TESTS
# =================================

n = len(df)

# Psychiatric: A more hallucinated -> Agentic better
k_psych = (df["Doc 1(Psychiatric) : Which summary has more hallucination?"] == "A").sum()
p_dom_psych = binomtest(k_psych, n, p=0.5, alternative="greater").pvalue

# Cardiopulmonary: B more hallucinated -> Agentic better
k_cardio = (df["Doc 2(Cardiopulmonary) : Which summary has more hallucination?"] == "B").sum()
p_dom_cardio = binomtest(k_cardio, n, p=0.5, alternative="greater").pvalue

# Oncology: A more hallucinated -> Vanilla better
k_onco = (df["Doc 3(Oncology): Which summary has more hallucination?"] == "A").sum()
p_dom_onco = binomtest(k_onco, n, p=0.5, alternative="greater").pvalue

print("\n=== Preference Dominance Tests (Binomial) ===")
print(f"Psychiatric     | correct={k_psych}/{n} | p={p_dom_psych:.6f}")
print(f"Cardiopulmonary | correct={k_cardio}/{n} | p={p_dom_cardio:.6f}")
print(f"Oncology        | correct={k_onco}/{n} | p={p_dom_onco:.6f}")

```
